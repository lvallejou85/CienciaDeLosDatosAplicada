# Ejercicio 2
Ejercicio.-- Algunos sistemas Unix tienen el comando tac (reverso de cat) el cual imprime el contenido de un archivo en inverso, es decir, empezando por la última línea. Implemente esta función en Python.

```python
def tac(archivo):
    f = open(archivo,'r').readlines()
    i=1
    while i < len(f):
        print(f[-i],end='')
        i+=1
    print(f[0],end='')
```

```python
tac('/etc/hosts')
```

Ejercicio.-- En Unix, el comando rev imprime el contenido de un archivo invirtiendo el orden de los caracteres de cada línea. Por ejemplo, si el archivo contiene las lineas 1234 y abcd, rev imprime 4321 y dcba. Implemente esta función en Python.

```python
def rev(archivo):
    f = open(archivo,'r').readlines()
    longstr=0
    i=0
    linea=''
    while i < len(f):
        j=0
        linearev = ''
        longstr = len(f[i]) - 1
        s = f[i]
        while j <= longstr:
            linearev += s[longstr - j]
            j+=1
        print(linearev,end='')
        i+=1
            
```

# Ejercicio 4
Ejercicio.-- Imprima un listado de valores únicos para el campo City del archivo files/customer.

```python
def ciudadesUnicas():

    x = open('customer','r').readlines()
    x = [z.replace('\n', '') for z in x]
    x = [z.split(',') for z in x]
    x = [z[2] for z in x[1:]]
    x = [z.replace('\"', '') for z in x]
    x = list(set(x))
    x.sort()
    for i in x:
        print(i)
```

# Ejercicio 6
Ejercicio.-- Agregue el campo Total a la tabla order. Este campo se calcula como la suma de ¿ProductID? * Price para cada OrderID en la tabla orderline.

```python
def TotalOrder():
    #Lee archivo de ordenes
    allord = open('allorders','r').readlines()
    allord = [z.replace('\n', '') for z in allord]
    allord = [z.split(',') for z in allord]
    
    #Lee archivo de detalle de órdenes
    orddet = open('orderdetail','r').readlines()
    orddet = [z.replace('\n', '') for z in orddet]
    orddet = [z.split(',') for z in orddet]

    #Crea una lista ordenada de las órdenes por el campo order.id
    allord_id = [z[1] for z in allord]
    orddet_id = [z[1] for z in orddet]
    allord_id.sort()
    orddet_id.sort()
    
    #Ordena las tablas por el campo order.id
    orddet.sort(key=lambda x:x[1])
    allord.sort(key=lambda x:x[1])
    
    posicion = 0
    posicion_det = 0
    
    for i in allord_id[1:]:
        
        if i in orddet_id[1:]:
            
            #posición de la orden en la tabla de detalles
            posicion = allord_id.index(i)
            posicion_det = orddet_id.index(i)
            
            suma_orden=0
            #obtiene la orden en la 'posicion'
            j = orddet_id[posicion_det]
            k = posicion_det
            while j == orddet_id[posicion_det]:
                
                suma_orden = suma_orden + (int(orddet[k][2]) * int(orddet[k][5]))
                
                #si no se ha terminado la lista
                if k+1 < len(orddet_id):
                    #Se obtiene la posición 'siguiente'
                    k = k + 1
                    j = orddet_id[k]
                    
                else:
                #Se terminó la lista
                    break
            
            allord[posicion] = allord[posicion] + [suma_orden]
            
    allord[0] = allord[0] + ['Total']
    print(allord[0:10])
```

```python
TotalOrder()
```
